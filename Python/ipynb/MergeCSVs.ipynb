```python
import os
import glob
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
os.chdir("/content/drive/My Drive/Test CSV Merge/")


extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
print(all_filenames[0])
print(pd.read_csv(all_filenames[0],encoding='utf-16', sep='\t') )
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f,encoding='utf-16', sep='\t') for f in all_filenames])

#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-16')
```
